```python
import pandas as pd
from datetime import date

fileList = []

#Cargar los archivos excel
def loadExcel () : 
  i = 1
  while i < 8:
    loadList = 'dm2019-01-0'+str(i)+'.xls'
    fileList.append(loadList)
    i += 1
loadExcel()
print(fileList)
```

```python
def loadDateTime () :
  i = 0
  while i < len(fileList):
    #Leer excel
    df = pd.read_excel(fileList[i])
    #Crear fecha
    createDate = df.iloc[1 , 0].split(",")[1].strip(' ').split(" ")
    createDate[1] = 1
    fixDate = date(int(createDate[2]) , createDate[1] , int(createDate[0]))
    #Crear columna y cargar fecha
    df['Fecha'] = pd.to_datetime(fixDate)
    fileList[i] = df
    i += 1
loadDateTime()
#Concatenar la data
#df = pd.concat(map(pd.DataFrame, fileList))
df = pd.concat(fileList)
```

```python
#Indique la cantidad de filas y columnas
print(f'Filas: {df.shape[0]}, Columnas: {df.shape[1]}\n')
#Indique los tipos de dato de las columnas.
print('Tipo de datos de las columnas')
print(df.dtypes)
```

```python
#DataFrame Original
display(df)
```

```python
#Limpiar los datos (sin destruir el dataframe original):
#Eliminar las filas a las que les faltan los datos numéricos (nan)
#Creamos una copia del DataFrame usando 'dropna()'
#datosMtDF = df.dropna()
datosMtDF = df.copy()
#Obtener columnas
getColumn = []
def loadColumns () :
  df = fileList[0]
  i = 0
  while i < df.shape[1]-1:
    selectColumn = df.iloc[3 , i]
    getColumn.append(selectColumn)
    i += 1
  getColumn.append('Fecha de muestra')
loadColumns()
print(getColumn)
print(len(getColumn))
#Renombrar las columnas más importantes.
i = 0
datosMtColumn = list(datosMtDF.columns)
while i < len(getColumn):
  datosMtDF = datosMtDF.rename(columns={datosMtColumn[i] : getColumn[i]})
  i += 1
datosMtDF = datosMtDF.drop(index=3)
#datosMtDF = datosMtDF.reset_index(drop=True)
#Código del profesor
#Para limpieza del dataset
estaciones_con_nan=datosMtDF[datosMtDF.isnull().any(axis=1)]['Estación'].unique()
datosMtDF.drop(datosMtDF[datosMtDF['Estación'].isin(estaciones_con_nan)].index, inplace=True)
print(f'Filas: {datosMtDF.shape[0]}, Columnas: {datosMtDF.shape[1]}\n')
```

```python
#Limpiar y convertir las columnas que deben ser numéricas
datosMtDF['Temperatura máxima (ºC)'] = datosMtDF['Temperatura máxima (ºC)'].str.split(" ").str[0].apply(pd.to_numeric)
datosMtDF['Temperatura mínima (ºC)'] = datosMtDF['Temperatura mínima (ºC)'].str.split(" ").str[0].apply(pd.to_numeric)
datosMtDF['Racha (km/h)'] = datosMtDF['Racha (km/h)'].str.split(" ").str[0].apply(pd.to_numeric)
datosMtDF['Velocidad máxima (km/h)'] = datosMtDF['Velocidad máxima (km/h)'].str.split(" ").str[0].apply(pd.to_numeric)
datosMtDF = datosMtDF.convert_dtypes()
print(datosMtDF.dtypes)
```

```python
#Analizar e indicar los siguientes puntos:
#Estadísticas count, mean, std, min, max, percentiles 25%, 50% y 75%.
display(datosMtDF.describe())
```

```python
#Medias por provincia.
display(datosMtDF.groupby(getColumn[1]).mean(numeric_only=True))
```

```python
#Mayores temperaturas máximas por estación.
print(datosMtDF.groupby(getColumn[0])[getColumn[2]].max().sort_values(ascending=False).head(5))
```

```python
#Menores temperaturas mínimas por estación.
display(datosMtDF.groupby([getColumn[12] , getColumn[0] , getColumn[1]])[getColumn[3]].min().sort_values(ascending=True).head(5))
```

```python
#Mayores rachas de viento en un día por provincia.
print(datosMtDF.groupby(getColumn[1])[getColumn[5]].max().sort_values(ascending=False).head(5))
```

```python
#Mostrar cuántas estaciones hay por provincia y un total general.
estacionProv = datosMtDF.groupby(['Provincia'])['Estación'].nunique()
print(estacionProv)
```
